```python
import nltk
newf = nltk.data.load('corpora/text_simplify_c/simple.txt')
test = newf
```

```python
tokens = nltk.word_tokenize(newf)
```

```python
result = ''.join([i for i in test if not i.isdigit()])      
```

```python
tokens = nltk.word_tokenize(result)
from collections import Counter
c = Counter(tokens)
```

```python
print(c)
```
