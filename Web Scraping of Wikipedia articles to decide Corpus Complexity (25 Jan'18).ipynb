```python
from bs4 import BeautifulSoup
import urllib3
import re
urllib3.disable_warnings()
http = urllib3.PoolManager()
url = "https://en.wikipedia.org/wiki/Main_Page"
response = http.request('GET', url)
soup = BeautifulSoup(response.data, "html5lib")
links = soup.find("div", {"id": "bodyContent"}).findAll("a", href=re.compile("(/wiki/)+([a-zA-Z0-9_:()])+"))
for link in links:
    print(link['href'])
```

```python
from bs4 import BeautifulSoup
import urllib3

import re
import nltk
from nltk.tokenize import word_tokenize

from collections import Counter

urllib3.disable_warnings()
http = urllib3.PoolManager()
url = ["https://en.wikipedia.org/wiki/Computer","https://simple.wikipedia.org/wiki/Computer"]
fin = []
for j in range(0,len(url)):
    x = 0
    response = http.request('GET', url[j])
    soup = BeautifulSoup(response.data, "html5lib")
    ps = soup.find("div", {"id": "bodyContent"}).findAll("p")
    cat_wiki = ''
    for i in range(0, len(ps)):
        cat_wiki += ps[i].text
    fin.append(cat_wiki)    


tok1 = word_tokenize(fin[0])
tok2 = word_tokenize(fin[1])
p_tok1 = nltk.pos_tag(tok1)
p_tok2 = nltk.pos_tag(tok2)
for x in range(0,len(tok1)):
    if tok1[x] == '(' or tok1[x] == ')' or tok1[x] == '.' or tok1[x] == ',' or tok1[x] == '[' or tok1[x] == ']' or tok1[x] == "'" or tok1[x] == '.':
        tok1[x] = ''      
for x in range(0,len(tok2)):
    if tok2[x] == '(' or tok2[x] == ')' or tok2[x] == '.' or tok2[x] == ',' or tok2[x] == '[' or tok2[x] == ']' or tok2[x] == "'" or tok2[x] == '.':
        tok2[x] = ''        
#print("S_Wiki : ", Counter(tok2))   
d_tok1 = dict(Counter(tok1))
d_tok2 = dict(Counter(tok2))
x = list(d_tok1.keys())
y = list(d_tok2.keys())
for i in range(0,len(x)):
    for j in range(0,len(y)):
        if x[i].lower() == y[j].lower() and x[i] != '':
            #print(x[i], d_tok1[x[i]], d_tok2[y[j]])
            lw = len(x[i])
            corpus_complexity = int(d_tok1[x[i]]) / int(d_tok2[y[j]])
            lexical_complexity = corpus_complexity * lw
            print(x[i], corpus_complexity, lexical_complexity)
#print(d_tok1)
```
